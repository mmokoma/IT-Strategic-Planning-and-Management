```python
# generate related variables
#data1=cases in last 24 hours, std=2367.162193, mean=610.260000
#data2=deaths in last 24 hours, std=25.245642, mean=8.380000

from numpy import mean
from numpy import std
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot
# seed random number generator
seed(1)
#preparing data
data1 = 2367.162193 * randn(50) + 610.260000
data2 = data1 + (25.245642 * randn(50) + 8.380000)
#print mean and stdv
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
#plotting
pyplot.scatter(data1, data2)
pyplot.show()
```

```python
np.cov(data1,data2)
```

The two variables go to same direction with value <b>5374292.23383622</b>

```python
#calculate the Pearson's correlation between two variables
from numpy.random import randn
from numpy.random import seed
from scipy.stats import pearsonr
import scipy.stats
#seed random number generator
seed(1)
#preparing the data
data1 = 2367.162193 * randn(50) + 610.260000
data2 = data1 + (25.245642 * randn(50) + 8.380000)
#calculate Pearson's correlation and the P-value
#corr, _ = pearsonr(data1, data2)
#print('Pearsons correlation: %.3f' % corr)
scipy.stats.pearsonr(data1, data2)
```

The correlation value is <b>0.9999629798059512</b> and the P-value is <b>8.435845820306244e-101</b>.
The two variables are positively correlated as the value is <b>0.9998652266773823</b> and it is close to <b>1</b>.
The Null hypothesis is to be accepted as the P-value is greater than <b>0.5</b>. Thefore, the deaths reported in last 24 hours are a result of cases reported in the last 24 hours.
